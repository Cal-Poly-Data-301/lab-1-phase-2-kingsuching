```python
from google.colab import files
import pandas as pd
responses = files.upload()
responses = pd.read_csv("response.csv")
responses
```

```python
import seaborn as sns
responses.columns
responses["accuracy"] = responses["Suppose you wanted to fill an average-sized school bus entirely with golf balls.  How many golf balls do you think you would need?"].map(lambda n: abs(n-1197360))
sns.violinplot(x=responses["How many siblings do you have?"], y=responses["accuracy"], hue=responses["Do you consider yourself a sports fan?"])
```

```python
sns.violinplot(x=responses["How many siblings do you have?"], y=responses["accuracy"], hue=responses["Do you enjoy video games?"])
```

```python
sns.violinplot(x=responses["How many siblings do you have?"], y=responses["accuracy"], hue=responses["Do you consider yourself artistic?"])
```

It is shown students who indicate that they are not sports fans tend to guess more accurately and consistently on the golf balls question depending on the number of siblings that they have.

In addition, people who play video games tend to do quite poorly if they don't have any siblings, according to the violin plot. However, for those that have siblings, the video games tend to have a positive correlation on how well the student does on the golf ball question.

Lastly, I noticed that the students who reported that they were artistic reported did really well on the golf ball question. For those that are not, their distributions are much more spread out which lowers the group accuracy as a whole.

```python
responses.drop("golfBalls", axis=1, inplace=True)
sns.pairplot(responses)
```

```python
pd.crosstab(responses["sports fan"], responses["artistic"], normalize=True).divide(responses["artistic"].value_counts(normalize=True)).plot.bar()
```

Given that a student reported that they were not artistic, they had a 51% chance of reporting that they were a sports fan. In turn, those that reported that they were artistic had only a 39% chance of reporting that they were sports fans. This illustrates how most students tended to be artistic *or* a sports fan, and not both.
